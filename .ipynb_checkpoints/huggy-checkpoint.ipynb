```python
img_height = 256
img_width =256
```

```python
def bytes_to_image(byte_data):
  try:
      bytes=byte_data['bytes']
      image=Image.open(BytesIO(bytes))
      
      if image.mode == 'P' or image.mode != 'RGB':
        print("Discarding image with mode:", image.mode)
        return None
      else:
        image = image.convert('RGB')
          
      image = image.resize((img_width, img_height))
      tensor_image = img_to_array(image)

      return tensor_image
  except (UnidentifiedImageError, IOError, ValueError) as e:
        print(f"Discarding corrupt or incompatible image: {e}")
        return None  # Return None for corrupt images
       
dataset['image']=dataset['image'].apply(bytes_to_image)
```

```python
dataset = dataset.dropna()
```

```python
def plot_image(index):
    np_image=(dataset['image'][index] * 255).astype('uint8')
    pil_image = Image.fromarray(np_image)
    plt.imshow(pil_image)
    plt.title(dataset['category'][index])
    plt.axis('off')
    plt.show()
```

```python
plot_image(67)
```

```python
X = dataset['image']
y = dataset['label']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=67)
```

```python
INPUT_SHAPE = (img_width,img_height,3)
num_classes = len(ds['train'].features['label'].names)

model = Sequential()
model.add(Rescaling(1/255))
model.add(Resizing(256,256))
model.add(RandomRotation(0.2))
model.add(RandomFlip('horizontal_and_vertical'))
model.add(Conv2D(32,(3,3),activation='relu',input_shape=INPUT_SHAPE))
model.add(Dropout(0.2))
model.add(MaxPooling2D())
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(Dropout(0.2))
model.add(MaxPooling2D())
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(Dropout(0.2))
model.add(MaxPooling2D())
model.add(Conv2D(num_classes,(3,3),activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_entropy',metrics=['accuracy'])
```
